### Q1. What is Flask Framework? What are the advantages of Flask Framework?
### Answer:
Flask is a micro web framework written in Python for building web applications without worrying about low-level details such as protocol, thread management,etc.<br>
<br>
The advantages of Flask Framework are as follows:<br>
> 1. Simplicity
> 2. Flexibility
> 3. Extensibility
> 4. Jinja2 templating
> 5. URL routing
> 6. RESTful support
> 7. Python based

### Q3. What is App routing in Flask? Why do we use app routes?
### Answer:
In Flask, app routing refers to the process of defining URL routes and associating them with specific view functions that handle incoming HTTP requests.<br>
<br>
We use app routes i.e. @app.route() a decorator, to map URL sent by client(browser) to corresponding view function (associated task for specific URL) of the Flask application to establish communication between client and server.Also it specify which HTTP methods (e.g., GET, POST, PUT, DELETE) a route should respond to by using the methods parameter in the @app.route() decorator. This ensures that flask application responds appropriately to different types of requests, such as reading data with GET or submitting data with POST.
